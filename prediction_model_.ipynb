```
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```
df=pd.read_csv("Cleaned_Sample_removed_cols.csv")
df.head()
```

```
# Convert FlightDate to datetime
df['FlightDate'] = pd.to_datetime(df['FlightDate'], format='%d-%m-%Y')
```

```
df.dtypes
```

```
# Drop duplicate rows
df.drop_duplicates(inplace=True)
```

```
df.drop(columns=['FlightDate'], inplace=True)
```

```
df.columns
```

```
# Remove the target column 'DepDel15' before encoding
df_droptarget = df.drop(columns=['DepDel15'])
```

```
from sklearn.preprocessing import OneHotEncoder
```

```
categorical_cols = ['Airline', 'Origin', 'Dest', 'Operating_Airline', 'OriginCityName', 'OriginStateName',
                    'DestCityName', 'DestState', 'DestStateName']

# Apply One-Hot Encoding to categorical columns, and drop the first column to avoid the dummy variable trap
df_encoded = pd.get_dummies(df_droptarget, columns=categorical_cols, drop_first=True)

# Checking the transformed dataframe
print(df_encoded.head())
```

```
df.head()
```

```
print(df_encoded.shape)  # Check the number of columns after encoding
```

```
from sklearn.preprocessing import StandardScaler

# Initialize the StandardScaler
scaler = StandardScaler()

# Fit and transform the data (excluding the target variable, which is already separated)
X_scaled = scaler.fit_transform(df_encoded)

# Check the scaled data (first few rows)
print(X_scaled[:5])
```

```
# Convert the scaled NumPy array back to a DataFrame
df_scaled = pd.DataFrame(X_scaled, columns=df_encoded.columns)

# Check the scaled DataFrame
print(df_scaled.head())
```

```
df_scaled.head()
```

```
import matplotlib.pyplot as plt
import numpy as np
from sklearn.decomposition import PCA

# Fit PCA to your data (df_encoded_features or df_reduced)
pca = PCA()
pca.fit(df_scaled)

# Calculate the cumulative explained variance
cumulative_variance = np.cumsum(pca.explained_variance_ratio_)

# Plot the cumulative explained variance
plt.figure(figsize=(10, 6))
plt.plot(range(1, len(cumulative_variance) + 1), cumulative_variance, marker='o', color='b')
plt.axhline(y=0.90, color='r', linestyle='--', label="90% Variance")
plt.axhline(y=0.95, color='g', linestyle='--', label="95% Variance")
plt.title('Cumulative Explained Variance vs. Number of Components')
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance')
plt.legend(loc='best')
plt.grid(True)
plt.show()
```

```
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# First, scale the one-hot encoded data
scaler = StandardScaler()
df_pca = scaler.fit_transform(df_scaled)

# Apply PCA to reduce dimensions
pca = PCA(n_components=680)  # Reduce to 100 dimensions, adjust as necessary
df_reduced = pca.fit_transform(df_pca)


```

```
# Check the shape after PCA
print(df_reduced.shape)
```

```
import seaborn as sns
import matplotlib.pyplot as plt

# Compute the correlation matrix of the reduced data
corr_matrix = pd.DataFrame(df_reduced).corr()

# Plot the correlation heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Correlation Matrix of PCA Components")
plt.show()
```

```
# Plot the correlation matrix of the top 10 components (or any number that is manageable)
top_n = 10  # Change to the number of components you want to inspect

corr_matrix_subset = pd.DataFrame(df_reduced).iloc[:, :top_n].corr()

# Plot the heatmap for the subset
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix_subset, annot=True, cmap='coolwarm', fmt='.2f')
plt.title(f"Correlation Matrix of Top {top_n} PCA Components")
plt.show()
```

```
from sklearn.model_selection import train_test_split

# Define your features (principal components)
X_reduced = df_reduced

# Define your target column (dependent variable)
y = df['DepDel15']

# Split into train and test sets (80% train, 20% test)
X_train, X_test, y_train, y_test = train_test_split(X_reduced, y, test_size=0.2, random_state=42)

# Check the shapes of the splits
print(f"Training data shape: {X_train.shape}, Test data shape: {X_test.shape}")
```

```
from sklearn.ensemble import RandomForestClassifier

# Initialize Random Forest Classifier
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)

# Train the model
rf_model.fit(X_train, y_train)

# Predict on the test data
y_pred_rf = rf_model.predict(X_test)


```

```
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```
# Evaluate the model
accuracy_rf = accuracy_score(y_test, y_pred_rf)
print(f"Random Forest Accuracy: {accuracy_rf}")
```

```
accuracy_percentage = accuracy_rf * 100
print(f"Random Forest Accuracy: {accuracy_percentage:.2f}%")
```
